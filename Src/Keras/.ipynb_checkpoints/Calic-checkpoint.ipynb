```python
import cv2
import numpy as np
import os
src_folder = "E:\\Image_Compressed\\Training_Data\\Original_old\\"
dst_folder = "E:\\Image_Compressed\\Training_Data\\Original\\"
fileNames = os.listdir(src_folder)
count = 0
for fileName in fileNames:
    image = cv2.imread(src_folder + fileName, cv2.IMREAD_GRAYSCALE)
    cv2.imwrite(dst_folder + fileName.split(".")[0] + ".pgm", image)       
    print(count)
    count += 1

# print(image.shape)
# cv2.imshow("screen", image[:,:,2])
# print(np.array_equal(image[:,:,0], image[:,:,1]) )
# cv2.waitKey(0)
# cv2.imwrite("test.pgm", image)
```

```python
import os
encoder = "/home/mj/HardDisk/Github/Image_Compressor/calic/encode/Calic_encoder.out"
decoder = "/home/mj/HardDisk/Github/Image_Compressor/calic/decode/Calic_decoder.out"
src_folder = "/home/mj/HardDisk/Github/Image_Compressor/Dataset/Validation_Data/Original/"
compressed_folder = "/home/mj/HardDisk/Github/Image_Compressor/Dataset/Validation_Data/Compressed/"
dst_folder        = "/home/mj/HardDisk/Github/Image_Compressor/Dataset/Validation_Data/Decompressed/"
fileNames = os.listdir(src_folder)

count = 0
for fileName in fileNames:
    os.system(encoder + " -in " + src_folder + fileName + " -out " + compressed_folder + fileName)
    os.system(decoder + " -in " + compressed_folder + fileName + " -out " + dst_folder + fileName)
    # print(decoder + " -in " + src_folder + fileName.split(".")[0] + ".calic" + " -out " + dst_folder + fileName)
    if count%1000 == 0:
        print(count)
    
    break
    count += 1
```
